https://www.gunviolencearchive.org/mass-shooting

The data features are the following:
* Incident ID: ID of the incident in the dataset;
* Incident Date: Date of the year when it has occured the incident;
* State: States in which the incident has happened;
* City Or Country: City or Country where the incident has happened;
* Address: Address where the incident has happened;
* Nr. Killed: Number of people that has been killed in the incident;
* Nr. Injured: Number of people that has been injured in the incident;
* Operations: Source of the single data.

## Data Manipulation
Here it will follow the manipulation of the data, where it will be done the following things:
* Joining the different dataframe & Re-index;
* Check the features type (like numeric, float, etc.);
* Check if there are empty values;
* Drop the column that are not going to be used;
* Re-order the data according to the date.

### Joining the dataframes & Re-Indexing

```python
# Joining the data
Mass_Shooting = pd.concat([ms2014,ms2015,ms2016,ms2017_2021])

# Re-Indexing
Mass_Shooting = Mass_Shooting.reset_index(drop=True)
```

```python
# CHaingin the data into format datetime
Mass_Shooting['Incident Date'] = pd.to_datetime(Mass_Shooting['Incident Date'])
```

**Comment**: It appear that there are empty values in the _Address_ colum and completely no value in the _Operations_ column. These two column, however, will not be used for this project.

### Drop the columns

The columns that are going to be dropped are:
* Incident ID;
* Address;
* Operations;

```python
Mass_Shooting = Mass_Shooting.drop(['Incident ID','Address','Operations'], axis=1)
```

## Re-order according to the date of the Incident

```python
# Ordering by date
Mass_Shooting = Mass_Shooting.sort_values(by='Incident Date')

# Re-Indexing
Mass_Shooting = Mass_Shooting.reset_index(drop=True)
```
